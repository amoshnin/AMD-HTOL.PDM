```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
from tqdm import tqdm
import numpy as np
```

```python
# Define path to folder containing CSV files
folder_path = './data'

# Get list of CSV files in the folder
xlsx_files = [file for file in glob.glob(os.path.join(folder_path, "*.xlsx")) if not os.path.basename(file).startswith("~$")]

# Sort CSV files based on date in the filename (assuming MM_DD_YY format)
xlsx_files.sort(key=lambda x: x.split("/")[-1].split(".")[0].split("_")[::-1])

# Create empty dataframe to store combined data
combined_df = pd.DataFrame()

# Read each CSV file and append to combined dataframe
for file in tqdm(xlsx_files, desc="Combining Files"):
    df = pd.read_excel(file, engine='openpyxl')

    time_col = 'time' if 'time' in df.columns else 'Time'

    # Convert 'time' column to datetime format
    df[time_col] = pd.to_datetime(df[time_col], format='%H:%M:%S.%f')

    # Combine 'Date' and time_col columns
    df['DateTime'] = df['Date'] + pd.to_timedelta(df[time_col].dt.time.astype(str))

    combined_df = pd.concat([combined_df, df], ignore_index=True)
```

```python
combined_df.to_csv('combined_data.csv')
```
